```python
from bs4 import BeautifulSoup
import requests
```

```python
html_text=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').text
```

```python
soup=BeautifulSoup(html_text,'lxml')
jobs=soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
jobs
```

```python
jobs_tag=soup.find('h3',class_='joblist-comp-name').text.replace('\r\n',' ')

jobs_tag
```

```python
print("Put some skill that you are not familiar with")
unfamiliar_skill=input('>')
print(f'Filtering out {unfamiliar_skill}')
```

```python
skills_tag=jobs.find('span',class_='srp-skills').text.replace('\n','').replace('\r',' ')
skills_tag
```

```python
published_date=jobs.find('span',class_='sim-posted').span.text
published_date
```

```python
for job in jobs:
    published_date=job.find('span',class_='sim-posted').span.text
    if 'few' in published_date:
        
        skills_tag=job.find('span',class_='srp-skills').text.replace('\n','').replace('\r',' ')
        more_info=job.header.h2.a['href']
        jobs_tag=job.find('h3',class_='joblist-comp-name').text.replace('\r\n',' ')
        if unfamiliar_skill not in skills_tag:
            print(f"Comapny name:{jobs_tag}")
            print(f"Required Skills:{skills_tag}")
            print(f"More info:{more_info}")
            print(' ')
```
